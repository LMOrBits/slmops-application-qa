```python
from litellm import completion
response = completion(
    model="openai/custom",               
    api_key="none",                  
    api_base="http://localhost:8080/v1",     
    messages=[
        {
            "role": "user",
            "content": "Hey, how's it going?",
        }
    ],
    stream=True
)
for part in response:
    print(part.choices[0].delta.content or "")
```
